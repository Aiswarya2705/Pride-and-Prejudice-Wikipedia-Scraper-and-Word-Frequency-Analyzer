# To find the most frequently used words in an article

```python
import nltk
from nltk.corpus import stopwords
from bs4 import BeautifulSoup
import urllib.request
import plotly.io as pio
```

## Read the page

```python
page=urllib.request.urlopen('https://en.wikipedia.org/wiki/Pride_and_Prejudice')
html_plain=page.read()
print(html_plain)
```

## Data Cleaning

```python
soup = BeautifulSoup(html_plain,'html.parser')
soup_text=soup.get_text(strip=True)
print(soup_text)
```

```python
# To avoid the machine recognizing the same words with different cases as different words due to their different Ascii values,
# convert all words to lower case
ready_text=soup_text.lower()
print(ready_text)
```

## Tokenization

```python
tokens = ready_text.split()
tokens[0:5]
```

### Cleaning up the stopwords

```python
import nltk
nltk.download('stopwords')
```

```python
stop_words=stopwords.words('english')
clean_tokens=tokens[:] # A copy of tokens

for token in tokens:
    if token in stop_words or not(token.isalnum()): # stopwords and special chars
        clean_tokens.remove(token)

print(clean_tokens)
```

## Data Visualization

```python
# Calculate the frequency of words using FreqDist function by NLTK
freq=nltk.FreqDist(clean_tokens)

for key,val in freq.items():
    print('Word: '+str(key)+', Quantity: '+ str(val))
```

```python
# Tokens used more than 10 times
high_freq=dict()
for key,val in freq.items():
    if (val>10):
        high_freq[key]=val
```

```python
fig=dict({
    "data":[{'type':'bar',
    "x":list(high_freq.keys()),
    "y":list(high_freq.values())}],
    "layout":{"title":{"text":"Most Frequently used words in 'Pride and Prejudice' wikipedia article "},
    "xaxis":{"categoryorder":"total descending"}}
})

pio.show(fig)
```
